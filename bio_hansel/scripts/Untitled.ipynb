```python
import pandas as pd
vcf_file="/home/CSCScience.ca/gtong/ecoli-genomes/core.vcf" 
with open(vcf_file, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('#CHROM'):
                break
cols = line.split()

df = pd.read_table(
        vcf_file, comment='#', header=None,
        names=cols).rename(columns={'#CHROM': 'CHROM'})
```

```python
df = df[df['ALT'].str.len() <= 1]
df = df.drop(['CHROM', 'ID', 'QUAL', 'FILTER', 'INFO', 'FORMAT'], 1)
```

```python
df.index=df.POS
sequence_data_frame=df[['REF', 'ALT']]
df=df.drop(['POS', 'REF', 'ALT'], 1)
df
```
